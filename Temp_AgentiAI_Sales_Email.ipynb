```python
from langchain.prompts import ChatPromptTemplate


from langchain_groq import ChatGroq
# used open-source model from Alibaba Cloud [google open source model --> gemma2-9b-it]
llm_groq = ChatGroq(model="gemma2-9b-it")
print()
resp1 = llm_groq.invoke("Hi, there")
print(f"Hi, I am opensource groq :- {resp1}")

llm_llama = ChatGroq(model="llama-3.1-8b-instant", temperature=0) # model="llama-3.3-70b-versatile"
resp2 = llm_llama.invoke("Hi, there")
print(f"Hi, I am opensource llama : {resp2}")

from langchain_openai import ChatOpenAI
llm_openai = ChatOpenAI(model="gpt-4o-mini")
resp3 = llm_openai.invoke("Hi, there!")
print(f"Hi, I am openai gpt-4o-mini :- {resp3}")


from langchain_google_genai import ChatGoogleGenerativeAI
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.2)
resp4 = llm_gemini.invoke("Hi, there!")
print(f"Hi, I am opensource gemini model :- {resp4}")
```

```python
from langchain_core.tools import Tool
from langchain.agents import initialize_agent, AgentType 


### Sales Agent 1
def cold_email_agent(query: str) -> str:
    instruction = """
    You are a sales agent working for ComplAI.
    You write professional, serious cold email for : {query}
    """
    return llm_gemini.invoke(instruction).content


tool1 = Tool(
    name="cold_email_writer",
    func=cold_email_agent,
    description="generate a cold email for sales"
)

sales_agent_1 = initialize_agent(
    name = "sales_agent_1",
    tools = [tool1],
    llm = llm_gemini,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
) 

response1 = sales_agent_1.invoke("Write a cold email about AI solution for finance industry")
print(response1)
```

```python

### Sales Agent 2
def witty_email_agent(query: str) -> str:
    instruction = """
    You are a humorous, engaging sales agent.
    You write witty, engaging cold emails that are likely to get a response for : {query}
    """
    
    return llm_llama.invoke(instruction).content


tool2 = Tool(
    name="witty_email_writer",
    func=witty_email_agent,
    description="generate a witty and engaging email for sales"
)

sales_agent_2 = initialize_agent(
    name = "sales_agent_2",
    tools = [tool2],
    llm = llm_llama,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
) 

response2 = sales_agent_2.invoke("Write a witty and engaging email about AI solution for finance industry")
print(response2)
```

```python

### Sales Agent 3
def concise_email_agent(query: str) -> str:
    instruction = """
    You are a busy sales agent.
    You write concise, to the point cold emails for : {query}
    """
    
    return llm_groq.invoke(instruction).content


tool3 = Tool(
    name="concise_email_writer",
    func=concise_email_agent,
    description="generate a concise and to the point email for sales"
)

sales_agent_3 = initialize_agent(
    name = "sales_agent_3",
    tools = [tool3],
    llm = llm_groq,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
) 

response3 = sales_agent_3.invoke("Write a concise and to the point email about AI solution for finance industry")
print(response3)
```

```python
# Sales Manager

def sales_manager(response1: str, response2:str, response3:str):
    
    sales_manager_instructions = f"""
    You are a Sales Manager. .
    You will be provided 3 reponses : {response1}, {response2} and {response3}.
    
    Your job is to pick the best one basis on your analysis.
 
    Follow these steps carefully:
    1. Generate Drafts: Use all three responses and generate three different email drafts. Do not proceed until all three drafts are ready.
 
    2. Evaluate and Select: Review the 3 drafts and choose the single best email using your judgment of which one is most effective.
    
 
    Crucial Rules:
    - Do not write them yourself.
    - Do not ignore the respones.
    """

    return llm_openai.invoke(sales_manager_instructions).content


manager_tool = Tool(
    name="sales manager",
    func = sales_manager,
    description="Your job is to access and analyse the 3 responses and pick the best one"
)

manager_agent = initialize_agent(
    tools = [manager_tool],
    llm = llm_openai,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)


manager_response = sales_manager(response1, response2, response3)
print(manager_response)

```

```python
### Subject Writer in HTML format

def html(manager_response: str):
    subject_instructions = f"""    
    You can convert a text email body to an HTML email body. 
    You are given a text email body - {manager_response} which might have some markdown and you need to convert it to an HTML email body with simple, clear, compelling layout and design
    """
    
    return llm_openai.invoke(subject_instructions).content

subject_tool = Tool(
    name="suject tool",
    func = html,
    description="Convert a text email body to an HTML email body"
)

html_agent = initialize_agent(
    tools = [subject_tool],
    llm = llm_openai,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)


final_response = html_agent.invoke(manager_response)
print(final_response)
```

```python
print(final_response['output'])
```

```python
### Subject Writer in Text format

def text(manager_response: str):
    subject_instructions = f"""    
    You have to give structured email. 
    You are given a text email body - {manager_response} which might have some markdown and you need to convert it to an text markdown so it can be seen proper email including subject, body email and closing sentence. layout and design
    """
    
    return llm_openai.invoke(subject_instructions).content

subject_tool_text = Tool(
    name="suject tool text",
    func = text,
    description="Convert a text email body to an structured email in markdown format"
)

text_agent = initialize_agent(
    tools = [subject_tool_text],
    llm = llm_openai,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)


final_response_text = text_agent.invoke(manager_response)
print(final_response_text["output"])
```
